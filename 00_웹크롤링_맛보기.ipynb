```python
# 크롤링 2 steps: get select
import requests
from bs4 import BeautifulSoup
```

```python
# 1단계 : get
# requests.get(url)
url = 'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EC%82%BC%EC%84%B1%EC%A0%84%EC%9E%90&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=40&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start=21'

resp = requests.get(url)
resp

# <Response [200]>: 응답코드가 같이 넘어오면, 200은 Success 정도의 의미이며, 2로 시작하면 괜찮
```

```python
# HTML 보기
# content 한글을 안보여줌 (decoding되지 않은 것)
# 텍스트로 보기보다는 content가 좋다. 오류 발생 예방

resp.content
```

```python
# text는 한글 보여줌 (decoding)

resp.text
```

```python
# 2단계 : select
# html.select(원하는 태그 표현 : 계층구조, 속성을 이용)
# html = resp.content
# soup = BeautifulSoup(html) # BS형으로 형변환 : select 함수를 적용하기위해
# 정규표현식: 문자열에서 내가 원하는 정보만 추출하고 싶을 때 사용 -> HTML에도 적용 가능

soup = BeautifulSoup(resp.content, 'lxml') # 파싱하는 것; HTML에 한해서는 마음대로 추출 가능; 정규표현식 불필요, lxml은 파싱 프로그램 이름, soup에 HTML 정보가 들어가게 됨

# 클래스가 news_tit인 a태그; soup에는 HTML이 BS형으로 들어가 있음
a_tags = soup.select('a.news_tit') # select하면 news_tit클래스를 모두 가져오는데, list로 들어간다.
# [<a_tag1~~~~~~~~>, <a_tag2~~~~~~~~~~~>, ....] for문으로 뽑자

# 이제 속성과 텍스트만 가져오면 된다.
# news_info = {}

titles, urls = [], [] # for문 앞에 빈 리스트 선언

for a_tag in a_tags:
    # TAG에서 속성만 지정하기 => dictionary처럼; BS형 변환에서 이렇게 만들어줌
    titles.append(a_tag['title'])
    # titles.append(a_tag.text) # 텍스트추춘
    urls.append(a_tag['href'])

news_info = {'제목' : titles,
             '링크' : urls}
news_info
```

```python
import pandas as pd

df_news = pd.DataFrame(news_info)
df_news.to_csv('삼성전자_기사정보(2023.5.8).csv',
               encoding='utf-8-sig', index=False) # utf-8-sig 인코딩 방식; 인코딩 방식에 따라 한글이 깨질 수도 있고 아닐 수도 있는데, utf-8-sig는 깨졌을 때 대부분 해결해준다.
# index=False; CSV에 인덱스 저장하지 않겠다.
df_news
```
